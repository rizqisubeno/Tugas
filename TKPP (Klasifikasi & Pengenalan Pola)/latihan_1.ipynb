```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
benda=np.array([[1.1, 1.3], [1.1, 1.5], [0.7, 1.5], [0.6, 1.9],
                [0.8, 2], [1.2, 0.6], [1.6, 0.5], [1.7, 0.7], 
                [2.2, 0.4], [2.2, 0.5], [2.3, 0.8], [1.5, 0.5]])
print(benda)
```

```python
lemari = benda[0:5]
buffet = benda[5:12]
# print("lemari : \n{}".format(lemari))
# print("buffet : \n{}".format(buffet))
```

```python
plt.scatter(x=lemari[:,0],y=lemari[:,1],color='darkorange',marker='o',label="lemari")
plt.scatter(x=buffet[:,0],y=buffet[:,1],color='brown',marker='P',label="buffet")
plt.xlabel("x1 = panjang")
plt.ylabel("x2 = tinggi")
```
