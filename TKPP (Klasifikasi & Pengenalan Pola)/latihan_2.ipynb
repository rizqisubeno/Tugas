```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
benda=np.array([[0.65, 0.76], [0.59, 0.81], [0.42, 0.91], [0.30, 0.95],
                [0.37, 0.93], [0.89, 0.45], [0.95, 0.30], [0.92, 0.38], 
                [0.98, 0.18], [0.98, 0.22], [0.94, 0.33], [0.95, 0.32],
                [0.71, 0.71], [0.63, 0.77], [0.71, 0.71], [0.84, 0.55]])
```

## Pisahkan x1=Panjang ; x2 = tinggi
1. ambil vektor ciri tiap kelas dengan mean tiap kelas

```python
x1 = benda[:,0]
x2 = benda[:,1]

kelas1 = benda[0:5,:]
kelas2 = benda[5:12,:]
kelas3 = benda[12:16,:]

ciri_kelas1 = kelas1.mean(axis=0)
ciri_kelas2 = kelas2.mean(axis=0)
ciri_kelas3 = kelas3.mean(axis=0)
```

2. buat fungsi untuk pengenalan barang

```python
def s1(arr, ciri_kelas1):
    return np.dot(arr, ciri_kelas1)
def s2(arr, ciri_kelas2):
    return np.dot(arr, ciri_kelas2)
def s3(arr, ciri_kelas3):
    return np.dot(arr, ciri_kelas3)
```

3. Kalkulasi dengan fungsi untuk pengenalan
4. deteksi barang berdasarkan nilai terbesar pada tiap kelas (s1, s2, s3)
5. buat tabel output

```python
out1 = s1(benda, ciri_kelas1)
out2 = s2(benda, ciri_kelas2)
out3 = s3(benda, ciri_kelas3)

df = pd.DataFrame({'x1':x1, 'x2':x2, 's1':out1, 's2':out2, 's3':out3})
# print index start from 1
df.index += 1
#print(df)

# check 
detect = np.empty(shape=len(df),dtype=object)
for i in df.index:
    if (df['s1'][i] > df['s2'][i]) and \
    (df['s1'][i] > df['s3'][i]):
        detect[i-1]="lemari"
    elif (df['s2'][i] > df['s1'][i]) and \
        (df['s2'][i] > df['s3'][i]):
        detect[i-1]="buffet"
    elif (df['s3'][i] > df['s1'][i]) and \
        (df['s3'][i] > df['s2'][i]):
        detect[i-1]="wardrope"
    else:
        detect[i-1]="Unknown"

#buat kolom baru untuk masukkan ke tabel baru
#s = pd.Series(detect)
df2 = pd.DataFrame({"deteksi": detect})
#print(s)
df["deteksi"] = df2["deteksi"].values
display(df)
```

6. Kalkulasi Akurasi

```python
num_error = 0
for i in df.index:
    if (i>=1 and i<=5) and (df['deteksi'][i]!="lemari"):
        num_error += 1
    elif (i>=6 and i<=12) and (df['deteksi'][i]!="buffet"):
        num_error += 1
    elif (i>=13 and i<=16) and (df['deteksi'][i]!="wardrope"):
        num_error += 1

akurasi = ((len(df)-num_error)/len(df))*100
print("akurasi deteksi adalah : {} %".format(akurasi))
```
